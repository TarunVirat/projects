```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

# Step 1: Make a request to the website
url = 'https://coal.gov.in/en/major-statistics/production-and-supplies'
response = requests.get(url)
html_content = response.content

# Step 2: Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Step 3: Find the table
table = soup.find('table')

# Step 4: Extract the table headers
headers = []
for header in table.find_all('th'):
    headers.append(header.text.strip())

# Step 5: Extract the table rows
rows = []
for row in table.find_all('tr')[1:]:
    cells = row.find_all('td')
    cells = [cell.text.strip() for cell in cells]
    rows.append(cells)

# Step 6: Create a DataFrame
df = pd.DataFrame(rows)

# Step 7: Display the DataFrame
rows[1]
rows[3]
```
